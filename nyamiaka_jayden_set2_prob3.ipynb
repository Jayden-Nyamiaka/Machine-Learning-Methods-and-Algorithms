# Problem 3 

Use this notebook to write your code for problem 3. 

```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/emiletimothy/Caltech-CS155-2023/main/set2/data/problem3data.txt', 'problem3data.txt')

train_data = np.loadtxt('problem3data.txt')
y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

## Lasso Regression

```python
DIM = 9
lasso_strengths = np.linspace(0.01, 3, 30)
lasso_weights = np.zeros((len(lasso_strengths), DIM))

# Trains a lasso model for each strength
idx = 0
for strength in lasso_strengths:
  lasso = Lasso(alpha=strength)
  lasso.fit(X_train, y_train)
  lasso_weights[idx] = lasso.coef_
  idx += 1

# Creates the figure
plt.figure()

# Plots each of the nine weight vectors against the strengths
for d in range(DIM):
  plt.plot(lasso_strengths, lasso_weights[:, d], marker='.')

# Labels the figure
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc='best')
plt.xlabel('Regularization Strength (α)')
plt.ylabel('Weight Value')
plt.title('Lasso Regression at Different Regularization Strengths')
```

## Ridge Regression

```python
DIM = 9
ridge_strengths = np.arange(1, 10001, 1)
ridge_weights = np.zeros((len(ridge_strengths), DIM))

# Trains a ridge model for each strength
idx = 0
for strength in ridge_strengths:
  ridge = Ridge(alpha=strength)
  ridge.fit(X_train, y_train)
  ridge_weights[idx] = ridge.coef_
  idx += 1

# Creates the figure
plt.figure()

# Plots each of the nine weight vectors against the strengths
for d in range(DIM):
  plt.plot(ridge_strengths, ridge_weights[:, d])

# Labels the figure
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc='best')
plt.xlabel('Regularization Strength (α)')
plt.ylabel('Weight Value')
plt.title('Ridge Regression at Different Regularization Strengths')
```
